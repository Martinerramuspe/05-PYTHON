```python
# PRIMERA forma de aplicarle una funcion a cada una de las celdas de una columna correspondiente a un dataframe. ES MAS ELEGANTE.
import pandas as pd
def algoritmo(x):
    return 4 + x

data = {'columna': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)

df['columna'].apply(algoritmo)
```

```python
# SEGUNDA  forma de aplicarle una funcion a cada una de las celdas de una columna correspondiente a un dataframe. ES MAS ROBUSTA.
import pandas as pd
def algoritmo(x):
    return  4 + x

data = {'columna': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)

for x in df['columna']:
   print (algoritmo(x))
```

```python
import re

# Definir la expresi√≥n regular
patron = re.compile(r'\b\w+\b')

# Buscar coincidencias en una cadena
cadena = "Esto es un ejemplo de expresiones regulares en Python."
resultados = patron.findall(cadena)

# Imprimir los resultados
print(resultados)
```
